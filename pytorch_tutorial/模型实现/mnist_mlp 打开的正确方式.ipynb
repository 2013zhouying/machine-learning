# MNIST with logsoft+NLLoss

```python
# 如果是多元分类，请用logsoftmax+NLLoss
```

```python
import torch
from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F
from torchvision import datasets,transforms
```

```python
batch_size = 64
```

```python
train_dataset = datasets.MNIST('data/',train=True,
                               transform=transforms.Compose([
                                   transforms.ToTensor(),
                                   transforms.Normalize((0.1307,),(0.3801,))
                               ]),
                               download=False)
```

```python
test_dataset = datasets.MNIST('data/',
                              transform=transforms.Compose([
                                  transforms.ToTensor(),
                                transforms.Normalize((0.1307,),(0.3801,))
                              ]),
                              download=False)
```

```python
train_loader = torch.utils.data.DataLoader(train_dataset,batch_size=batch_size,shuffle=True)
test_loader = torch.utils.data.DataLoader(train_dataset,batch_size=batch_size,shuffle=True)
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        self.l1 = nn.Linear(784,520)
        self.l2 = nn.Linear(520,320)
        self.l3 = nn.Linear(320,240)
        self.l4 = nn.Linear(240,120)
        self.l5 = nn.Linear(120,10)
        self.relu = nn.ReLU()
        self.logsoftmax = nn.LogSoftmax()
    def forward(self, x):
        x = x.view(-1,784)
        out = self.relu(self.l1(x))
        out = self.relu(self.l2(out))
        out = self.relu(self.l3(out))
        out = self.relu(self.l4(out))
        out = self.relu(self.l5(out))
        return out
```

```python
net = Net()
```

```python
#  在这里有一个注意点 就是多元分类使用的是交叉熵
#  如果输出的为softmax 则使用的损失函数为负对数损失函数
#  如果输出的为线性的 则使用交叉熵损失函数
#  也就是 softmax+nlloss =  linear + CrossEntropyLoss


loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(net.parameters(),lr=0.01, momentum=0.5)
```

```python
for epochs in range(2):
    for batch_index, (data,target) in enumerate(train_loader):
        data, target = Variable(data), Variable(target)
        
        y_pred = net(data)
        loss = loss_fn(y_pred,target)
        
        print(epochs,batch_index,loss.data[0])
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
```

```python
correct = 0
for data,target in test_loader:
    net.eval()
    data, target = Variable(data), Variable(target)
    y_pred = net(data)
    pred = y_pred.data.max(1)[1]
    correct += (target.data==pred).sum()
print(correct/len(test_loader.dataset))
```

```python
net.eval()
test_loss = 0
correct = 0
for data, target in test_loader:
    data, target = Variable(data, volatile=True), Variable(target)
    output = net(data)
    # get the index of the max log-probability
    pred = output.data.max(1, keepdim=True)[1]
    correct += pred.eq(target.data.view_as(pred)).cpu().sum()

test_loss /= len(test_loader.dataset)
print('({:.0f}%)'.format(100. * correct / len(test_loader.dataset)))
```
