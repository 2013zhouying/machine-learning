# diabets detection

```python
# 首先，该问题也是一个二元分类模型，即使用的损失函数为BCELoss

import torch
from torch.autograd import Variable
import torch.nn as nn
import numpy as np
```

```python
# load data
xy = np.loadtxt('data-diabetes.csv',delimiter=',',dtype=np.float32)
x_data = Variable(torch.from_numpy(xy[:,:-1]))
y_data = Variable(torch.from_numpy(xy[:,[-1]]))
```

```python
print(x_data.data.shape)
print(y_data.data.shape)
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        self.linear1 = nn.Linear(8,6)
        self.linear2 = nn.Linear(6,4)
        self.linear3 = nn.Linear(4,1)
        self.sigmoid = nn.Sigmoid()
    def forward(self,x):
        out = self.sigmoid(self.linear1(x))
        out = self.sigmoid(self.linear2(out))
        out = self.sigmoid(self.linear3(out))
        return out
```

```python
net = Net()
loss_fn = nn.BCELoss(size_average=True)
```

```python
optimizer = torch.optim.Adam(net.parameters(),lr=0.001)
```

```python
for t in range(50000):
    y_pred = net(x_data)
    loss = loss_fn(y_pred,y_data)
    print(t,loss.data[0])
    
    optimizer.zero_grad()
    
    loss.backward()
    
    optimizer.step()
```
