# 逻辑回归模型

```python
import torch
from torch.autograd import Variable
import torch.nn as nn
import numpy as np
```

```python
x_data = Variable(torch.Tensor([[1.0,1.0], [0.,0.], [1.,0.], [0.,1.]]))
y_data = Variable(torch.Tensor([[0.], [0.], [1.], [1.]]))
```

```python
class Model(nn.Module):
    def __init__(self):
        super(Model,self).__init__()
        self.linear1 = nn.Linear(2,4)
        self.linear2 = nn.Linear(4,1)
    def forward(self,x):
        out = torch.nn.functional.sigmoid(self.linear1(x))
        out = torch.nn.functional.sigmoid(self.linear2(out))
        return out
```

```python
net = Model()
```

```python
loss_fn = nn.BCELoss()

#  注意 一定要使用BCELoss 该损失函数是针对于二元分类的情况，也就是标签只能为0或1
optimizer = torch.optim.Adam(net.parameters(),lr=1e-2)
```

```python
for t in range(10000):
    y_pred = net(x_data)
    loss = loss_fn(y_pred,y_data)
    print(t,loss.data[0])
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
test = Variable(torch.Tensor([[1.,0.]]))
print(net.forward(test).data[0][0])
```
